```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
# Uploading data frame 1
df1 = pd.read_csv('imdb.title.ratings.csv.gz')
df1
```

```python
# To look for null values
df1.info()
```

```python
# To show basic statistics
df1.describe()
```

```python
# Uploading data frame 2
df2 = pd.read_csv('imdb.title.basics.csv.gz')
df2
```

```python
# Uploading data frame 3
df3 = pd.read_csv('bom.movie_gross.csv.gz')
df3
```

```python
# Merging data frames 1 and 2 using 'inner join'.

merged_df = pd.merge(df2,df1,on='tconst',how='inner')
merged_df
```

```python
# Renaming column name for merging merged_df to df3
merged_df.rename(columns={'primary_title': 'title'}, inplace=True)
merged_df
```

```python
# Merging data frame 3 with data frames 1 and 2 using 'title' as primary key and using 'outer join'.

merged_df1 = pd.merge(df3,merged_df,on='title',how='outer')
merged_df1
```

```python
# Dropping original_title in the columns list
merged_df1 = merged_df1.drop('original_title', axis=1) 
merged_df1
```

```python
# To look for duplicate values
merged_df1.duplicated().value_counts()
```
