```python
# Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
%matplotlib inline
```

```python
# Uploading data frame 1
df1 = pd.read_csv('imdb.title.ratings.csv.gz')
df1
```

```python
# To look for null values
df1.info()
```

```python
# To show basic statistics
df1.describe()
```

```python
# Uploading data frame 2
df2 = pd.read_csv('imdb.title.basics.csv.gz')
df2
```

```python
# Uploading data frame 3
df3 = pd.read_csv('bom.movie_gross.csv.gz')
df3
```

```python
# Merging data frames 1 and 2 using 'inner join'.

merged_df = pd.merge(df2,df1,on='tconst',how='inner')
merged_df
```

```python
# Renaming column name for merging merged_df to df3
merged_df.rename(columns={'primary_title': 'title'}, inplace=True)
merged_df
```

```python
# Merging data frame 3 with data frames 1 and 2 using 'title' as primary key and using 'left join'.

merged_df1 = pd.merge(df3,merged_df,on='title',how='left')
merged_df1
```

```python
# Dropping original_title in the columns list
merged_df1 = merged_df1.drop('original_title', axis=1) 
merged_df1
```

```python
# To look for duplicate values
merged_df1.duplicated().value_counts()
```

```python
# Looking for missing values
merged_df1.isna().sum()
```

```python
# To look for statistics
merged_df1.describe()
```

```python
# To look for null values
merged_df1.info()
```

```python
# Capitalizing the name of columns
merged_df1.rename(columns={'title':'Title','studio':'Studio','domestic_gross':'Domestic_gross','foreign_gross':'Foreign_gross','tconst':'Tconst','start_year':'Start_year','runtime_minutes':'Runtime_minutes','genres':'Genres','averagerating':'Average_rating','numvotes':'Number_of_votes'}, inplace=True)
```

```python
# Dropping year in the columns list
merged_df1 = merged_df1.drop('year', axis=1) 
merged_df1
```

```python
#drops rows with missing values
merged_df1.dropna()
```

```python
#Create data frame
df = pd.DataFrame(merged_df1)

# Splitting genres and duplicating rows
df_genres = df['Genres'].str.split(',', expand=True).stack().reset_index(level=1, drop=True)
df_genres.name = 'Genre'
df = df.drop('Genres', axis=1).join(df_genres)

# Grouping by genre and calculating average rating
genre_avg_rating = df.groupby('Genre')['Average_rating'].mean().sort_values()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
genre_avg_rating.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Rating by Genre')
plt.xlabel('Genre')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Upload data frame
df = pd.DataFrame(merged_df1)

# Splitting genres and duplicating rows
df_genres = df['Genres'].str.split(',', expand=True).stack().reset_index(level=1, drop=True)
df_genres.name = 'Genre'
df = df.drop('Genres', axis=1).join(df_genres)

# Grouping by genre and calculating total domestic sales
genre_domestic_sales = df.groupby('Genre')['Domestic_gross'].sum().sort_values()

# Plotting the bar chart
plt.figure(figsize=(10, 6))
genre_domestic_sales.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Total Domestic Sales by Genre')
plt.xlabel('Genre')
plt.ylabel('Total Domestic Sales')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
```

```python
# Upload data frame
df = pd.DataFrame(merged_df1)

# Drop rows with missing ratings
df.dropna(subset=['Average_rating'], inplace=True)

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Average_rating'], bins=10, color='skyblue', edgecolor='black')
plt.title('Frequency Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Upload data frame
df = pd.DataFrame(merged_df1)

# Replace NaN with an integer (for example, 0)
df.fillna(0, inplace=True)

# Splitting genres and expanding rows
df_genres = df['Genres'].str.split(',', expand=True).stack().reset_index(level=1, drop=True)

# Adding the genres as a column
df = df.drop('Genres', axis=1).join(df_genres.rename('Genre'))

# Group by genre and sum foreign gross
df['Foreign_gross'] = pd.to_numeric(df['Foreign_gross'], errors='coerce')
genre_foreign_gross = df.groupby('Genre')['Foreign_gross'].sum().sort_values(ascending=True)

# Plotting
plt.figure(figsize=(10, 6))
genre_foreign_gross.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Foreign Gross by Genre')
plt.xlabel('Genre')
plt.ylabel('Foreign Gross (in billions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
